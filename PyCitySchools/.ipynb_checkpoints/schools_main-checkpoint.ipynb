# School Data Trends

65% of the nearly 40,000 students in the district passed both their math and reading test. Looking at the test data broken down by the 15 schools in the district revealed a few trends in what schools had more students pass the tests. The percent of students passing the math or reading test very closely followed the trends seen with the percent of students passing both tests, so only the overall percent passing rate will be discussed for brevity. 
Smaller schools, defined as fewer than 2,000 students, had an average passing rate of 90%, while medium or large schools had passing rates of 63% and 54%. Charter schools had a 90% pass rate while district schools had a much lower pass rate of 54%. These two trends could be consistent with a philosophy that additional and specialized attention results in better test scores for students. Schools that had the smallest budget per student had 90% of students pass, on average, while the schools spent the most per student had 59% pass on average. Somewhat counter-intuitively spending more per student correlated with lower percent passing rates. This could be investigated further by breaking the budget of schools to see if a particular type of spending is correlated with a lower percent of students passing.

```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#Clean Up Headers
school_data_df = school_data.rename(columns={"school_name": "school name"})
student_data_df = student_data.rename(columns={"school_name": "school name", "student_name": "student name", "reading_score": "reading score", "math_score": "math score"})

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data_df, school_data_df, how="left", on=["school name", "school name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calc descriptive variables
students = school_data_complete["student name"].count()
schools = school_data_complete["school name"].nunique()
total_budget = school_data_df["budget"].sum()
avg_math_score = round(school_data_complete["math score"].mean())
avg_reading_score = round(school_data_complete["reading score"].mean())
avg_overall_score = round((float(avg_math_score)+float(avg_reading_score))/2)
```

```python
#Calc percent passing
school_data_reading_pass = school_data_complete.loc[school_data_complete["reading score"] >= 70]
school_data_math_pass = school_data_complete.loc[school_data_complete["math score"] >= 70]
percent_pass_math = round(float(school_data_math_pass["student name"].count())/float(students)*100, 2)
percent_pass_reading = round(float(school_data_reading_pass["student name"].count())/float(students)*100,2)

school_data_overall_pass = school_data_reading_pass.loc[school_data_complete["math score"] >= 70]
percent_pass_overall = round(float(school_data_overall_pass["student name"].count())/float(students)*100,2)
```

```python
summary_stats_count_df = pd.DataFrame({"Students": [students], "Schools": [schools], "Budget": [total_budget]})
summary_stats_count_df = summary_stats_count_df.rename(index={0: "Total Number"})
summary_stats_count_df
```

```python
summary_stats_score_df = pd.DataFrame({"Average Score": [avg_math_score, avg_reading_score, avg_overall_score], "Percent of Students Passing": [percent_pass_math, percent_pass_reading, percent_pass_overall]}, index=['Math', 'Reading', 'Overall'])
summary_stats_score_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_data_overview = school_data_df[["school name", "type", "size", "budget"]]
school_data_overview
```

```python
school_data_overview = school_data_df[["school name", "type", "size", "budget"]]
school_data_overview["Budget per Student"] = round(school_data_overview["budget"]/school_data_overview["size"], 2)

school_grouped_complete = school_data_complete.groupby("school name")
school_score_avg = round(school_grouped_complete[["math score", "reading score"]].mean(), 2)
school_score_avg=school_score_avg.rename(columns={"math score": "average math score", "reading score": "average reading score"})
```

```python
#Number Passed 
school_grouped_reading_pass=school_data_reading_pass.groupby("school name")
school_pass_reading_df = school_grouped_reading_pass["student name"].count().to_frame()
school_pass_reading_df = school_pass_reading_df.rename(columns={"student name": "total students pass reading"})

school_grouped_math_pass=school_data_math_pass.groupby("school name")
school_pass_math_df = school_grouped_math_pass["student name"].count().to_frame()
school_pass_math_df = school_pass_math_df.rename(columns={"student name": "total students pass math"})

school_grouped_overall_pass=school_data_overall_pass.groupby("school name")
school_pass_overall_df = school_grouped_overall_pass["student name"].count().to_frame()
school_pass_overall_df = school_pass_overall_df.rename(columns={"student name": "total students pass both"})

#Merge counts
school_pass_df = pd.merge(school_pass_reading_df, school_pass_math_df, on="school name")
school_pass_df = pd.merge(school_pass_df, school_pass_overall_df, on="school name")

#collecting school data
school_grouped_complete_df = pd.merge(school_data_overview, school_pass_df, on="school name")
school_grouped_complete_df = pd.merge(school_grouped_complete_df, school_score_avg, on="school name")

#calc percent
school_grouped_complete_df["percent pass math"] = round(school_grouped_complete_df["total students pass math"]/school_grouped_complete_df["size"]*100, 2)
school_grouped_complete_df["percent pass reading"] = round(school_grouped_complete_df["total students pass reading"]/school_grouped_complete_df["size"]*100, 2)
school_grouped_complete_df["percent pass both"] = round(school_grouped_complete_df["total students pass both"]/school_grouped_complete_df["size"]*100, 2)

school_grouped_complete_df
```

## Top Performing Schools (By % Overall Passing)

```python
school_grouped_complete_df.sort_values("percent pass both", ascending=False).head()
```

## Bottom Performing Schools (By % Overall Passing)

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Binning
budget_bins = [575, 595, 615, 635, 655]

# Create the names for the bins
bin_names = ['$575 - 595', "$596 - 615","$616 - 635","$636 - 655"]
school_data_overview.sort_values("Budget per Student")

#add bin column
school_grouped_complete_df["budget group"] = pd.cut(school_grouped_complete_df["Budget per Student"], budget_bins, labels=bin_names, include_lowest=True)

#group by and display
school_grouped_complete_budget = round(school_grouped_complete_df.groupby(["budget group"]).agg({'average reading score': ['mean'], 'average math score': ['mean'], 'percent pass reading': ['mean'],  'percent pass math': ['mean'],  'percent pass both': ['mean']}), 2)

#percent calc
school_grouped_complete_budget
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Binning
size_bins = [0, 1999, 3000, 5000]

# Create the names for the bins
size_bin_names = ["Small, less than 2,000", "Medium, 2,000 to 3,000","Large, greater than 3,000"]
school_data_overview.sort_values("size")

#add bin column
school_grouped_complete_df["size group"] = pd.cut(school_grouped_complete_df["size"], size_bins, labels=size_bin_names, include_lowest=True)
school_grouped_complete_df

#group by and display
school_grouped_complete_size= round(school_grouped_complete_df.groupby(["size group"]).agg({'average reading score': ['mean'], 'average math score': ['mean'], 'percent pass reading': ['mean'],  'percent pass math': ['mean'],  'percent pass both': ['mean']}), 2)

school_grouped_complete_size
```
